```python
import cv2
import os
```

```python
default_dir = "C:/anoosheh/CP/Master/Thesis/Dataset/Dataset_under_Delta_robot_without_light_0_degree/Videos"
input_dir = default_dir + "/videos"
```

```python
for i in os.listdir(default_dir):
  video_path = os.path.join(default_dir, i)
  vcap = cv2.VideoCapture(video_path)
  cnt = 0
  while True:
    _, frame = vcap.read()
    if frame is None:
      break

    if cnt%5 == 0:
      cv2.imwrite(default_dir + "/frames/frame_"+ i[:-4] + "_" + str(cnt) + ".jpg", frame)
    cnt += 1
cv2.destroyAllWindows()
```
